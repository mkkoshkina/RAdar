```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
# Load the SNP catalog 
# source: https://www.pgscatalog.org/score/PGS000195/
# source article: https://pubmed.ncbi.nlm.nih.gov/36514807/
df = pd.read_csv("PGS000195_hmPOS_GRCh37.txt", sep="\t", comment='#')
df
```

```python
#filter for SNPs with rsID
df.dropna(subset=['rsID'], inplace=True)
df
```

```python
#check if the rsID are the same for GRCh37 and GRCh38
df[df['rsID']!=df['hm_rsID']]
```

```python
#load GWAS catalog data
df_gwas_catalog = pd.read_csv('gwas_catalog_v1.0-associations_e114_r2025-07-21.tsv', sep='\t')
df_gwas_catalog
```

```python
df_gwas_catalog.columns
```

```python
#rename columns for subsequent merge with SNPs from the PGS catalog
df_gwas_catalog = df_gwas_catalog.rename(columns={'SNPS': 'rsID', 'CHR_ID': 'chr_name', 'CHR_POS': 'chr_position'})
```

```python
df_gwas_catalog.dropna(subset=['chr_name'], inplace=True)
df_gwas_catalog.dropna(subset=['chr_position'], inplace=True)
```

```python
df_gwas_catalog['chr_name'].unique()
```

```python
#filter for autosomal chromosomes only
df_gwas_catalog = df_gwas_catalog[df_gwas_catalog['chr_name'].isin([str(i) for i in range(1, 23)])]
```

```python
df_gwas_catalog['chr_name'] = df_gwas_catalog['chr_name'].astype(int)
df_gwas_catalog['chr_position'] = df_gwas_catalog['chr_position'].astype(int)
df['chr_position'] = df['chr_position'].astype(int)
```

```python
#merging the data with gwas catalog. We could  simply use the rsID as they should be unique,
#but we also take into account the genomic position to ensure accuracy.
dfdf_gwas_merge = df.merge(df_gwas_catalog, on=['rsID', 'chr_name','chr_position'], how='left')
dfdf_gwas_merge
```

```python
#we didn't lose any mutations after the merge
len(dfdf_gwas_merge['rsID'].unique())
```

```python
dfdf_gwas_merge.columns
```

```python
dfdf_gwas_merge = dfdf_gwas_merge[['rsID', 'chr_name', 'chr_position', 'effect_allele', 'other_allele',
       'effect_weight', 'OR', 'LINK', 'STUDY', 'DISEASE/TRAIT', 'MAPPED_GENE', 'SNP_GENE_IDS']]
```

```python
dfdf_gwas_merge = dfdf_gwas_merge.drop_duplicates()
```

```python
#small function to get gene information from Ensembl using the REST API
import requests

def get_gene_info(ensembl_id):
    url = f"https://rest.ensembl.org/lookup/id/{ensembl_id}?expand=1"
    headers = {"Content-Type": "application/json"}

    response = requests.get(url, headers=headers)

    if response.ok:
        data = response.json()
        return {
            "ensembl_id": ensembl_id,
            "gene_name": data.get("display_name"),
            "description": data.get("description"),
            "biotype": data.get("biotype"),
            "species": data.get("species")
        }
    else:
        return {"ensembl_id": ensembl_id, "gene_name": None, "description": None}

# Example:
gene_id = "ENSG00000142606"
info = get_gene_info(gene_id)
print(info)
```

```python
#some genes have multiple Ensembl IDs, we need to split them
dfdf_gwas_merge['SNP_GENE_IDS'] = dfdf_gwas_merge['SNP_GENE_IDS'].apply(
    lambda x: x.split(',') if isinstance(x, str) and ',' in x else ([x] if pd.notna(x) else x)
)

dfdf_gwas_merge
```

```python
dfdf_gwas_merge = dfdf_gwas_merge.explode('SNP_GENE_IDS')
dfdf_gwas_merge
```

```python
#quering the ensemble ids list
results = {}

for id in [str(x).strip() for x in list(set(dfdf_gwas_merge['SNP_GENE_IDS'].tolist()))]:
    res = get_gene_info(id)
    results[id] = res['description']
```

```python
results
```

```python
import json

# Save to JSON file
with open("gene_mapping.json", "w") as f:
    json.dump(results, f, indent=2)
```

```python
dfdf_gwas_merge['GENE_DESCRIPTON'] = dfdf_gwas_merge['SNP_GENE_IDS'].apply(lambda x: results[x.strip()] if pd.notna(x) else x)
```

```python
dfdf_gwas_merge['GENE_DESCRIPTON'] = dfdf_gwas_merge['GENE_DESCRIPTON'].astype(str)
dfdf_gwas_merge['gene_description'] = dfdf_gwas_merge['GENE_DESCRIPTON'].apply(lambda x: x[:x.index('[Source')] if '[Source' in x else x)
```

```python
tags = ['Arthritis', 'Rheumatoid arthritis', 'arthritis', 'rheumatoid arthritis', 'RA', 'rheumatoid', 'Autoimmune traits']
```

```python
dfdf_gwas_merge['tokeep_1'] = dfdf_gwas_merge['DISEASE/TRAIT'].apply(
    lambda x: any(tag in str(x) for tag in tags)
)
dfdf_gwas_merge['tokeep_2'] = dfdf_gwas_merge['STUDY'].apply(
    lambda x: any(tag in str(x) for tag in tags)
)
```

```python
dfdf_gwas_merge = dfdf_gwas_merge[['rsID', 'chr_name', 'chr_position', 'effect_allele', 'other_allele',
       'effect_weight', 'OR', 'LINK', 'STUDY', 'DISEASE/TRAIT', 'MAPPED_GENE',
       'SNP_GENE_IDS', 'gene_description']]
```

```python
import numpy as np
```

```python
def first_not_nan(x):
    return x.dropna().iloc[0] if x.notna().any() else np.nan
```

```python
#collapsing all the information for the same rsID in one row
dfdf_gwas_merge = dfdf_gwas_merge.groupby('rsID').agg({
    'LINK': lambda x: ';'.join(map(str, x)),
    'STUDY': lambda x: ';'.join(map(str, x)),
    'DISEASE/TRAIT': lambda x: ';'.join(map(str, x)),
    'chr_name': 'first',
    'chr_position': 'first',
    'effect_allele': 'first',
    'other_allele': 'first',
    'effect_weight': 'first',
    'OR': 'first',
    'MAPPED_GENE': first_not_nan,
    'SNP_GENE_IDS': first_not_nan,
    'gene_description': first_not_nan
}).reset_index()
dfdf_gwas_merge
```

```python
dfdf_gwas_merge = dfdf_gwas_merge.rename(columns={'LINK': 'Sources', 'chr_name': 'Chromosome', 'chr_position': 'Position',
       'effect_allele': 'Effect allele', 'other_allele': 'Other allele', 'effect_weight': 'Effect weight', 'OR': 'Odds ratio', 'MAPPED_GENE': 'Gene symbol',
       'SNP_GENE_IDS': 'Ensembl gene ID', 'gene_description': 'Gene description'})
```
